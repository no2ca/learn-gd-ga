```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def obj_func(x):
    return (x - 3)**2 + 1
```

```python
def gradient(x):
    return 2 * (x - 3)
```

```python
def gd(start_x, learning_rate, max_iter, tolerance=1e-4):
    x = start_x
    x_history = [x]
    f_history = [obj_func(x)]

    for i in range(max_iter):
        grad = gradient(x)
        x_new = x - learning_rate * grad
        x_history.append(x_new)
        f_history.append(obj_func(x_new))

        if abs(x_new - x) < tolerance:
            print(f"収束しました(反復回数: {i+1})")
            print(f"収束した値: {x_new}")
            break
        elif i == max_iter - 1:
            print(f"収束しませんでした")
            print(f"値: {x_new}")

        x = x_new

    return np.array(x_history), np.array(f_history)
```

```python
def visualize(start_x, learning_rate, max_iter):
    x_history, f_history = gd(start_x, learning_rate, max_iter)

    x_plot = np.linspace(-1, 6, 1000)
    y_plot = obj_func(x_plot)

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 1, 1)
    plt.plot(x_plot, y_plot)
    plt.plot(x_history, f_history, '-ro', markersize=8)

    plt.subplot(2, 1, 2)
    iterations = range(len(f_history))
    plt.plot(iterations, f_history, '-bo')
    plt.axhline(y=1.0, color='r', linestyle='--', label='min_value = 1.0')
    plt.xlabel('iteration rate')
    plt.ylabel('value of f(x)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks(range(len(f_history)))
    
    plt.show()

visualize(0.0, 0.2, 50)
```
